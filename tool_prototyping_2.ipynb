```python
def parse_cigar(cigar):
    left_clip = 0; right_clip = 0
    num = ''
    op_num = 0
    for char in cigar:
        if char.isalpha():
            if char in 'HS':
                if op_num == 0:
                    left_clip += int(num)
                else:
                    right_clip += int(num)
            num = ''
            op_num += 1
        else:
            num += char
    return (left_clip, right_clip)
```
